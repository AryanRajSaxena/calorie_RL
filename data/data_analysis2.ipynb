```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('meal3_cleaned.csv')
df
```

```python
# Define mappings for meal_time categories
breakfast_keywords = ['breakfast', 'Breakfast', 'Breakfast/Snack', 'breakfast/snacks', 'breakfast/lunch']
lunch_keywords = ['lunch', 'Lunch', 'Lunch/Dinner', 'Lunch/Snack']
dinner_keywords = ['dinner', 'Dinner', 'Lunch/Dinner']
snacks_keywords = ['snacks', 'Snack', 'Snacks', 'Dessert', 'Dessert/Snacks', 'Beverage/Snacks', 'Post-Workout', 'Side Dish', 'salad', 'side dish', 'beverage']
desert_keywords = ['dessert', 'Dessert', 'Dessert/Snacks']

# Function to map meal_time to categories
def map_meal_time(meal_time):
    categories = []
    if isinstance(meal_time, str):  # Check if meal_time is a string
        meal_time_lower = meal_time.lower()
        if any(keyword in meal_time_lower for keyword in breakfast_keywords):
            categories.append('breakfast')
        if any(keyword in meal_time_lower for keyword in lunch_keywords):
            categories.append('lunch')
        if any(keyword in meal_time_lower for keyword in dinner_keywords):
            categories.append('dinner')
        if any(keyword in meal_time_lower for keyword in snacks_keywords):
            categories.append('snacks')
        if any(keyword in meal_time_lower for keyword in desert_keywords):
            categories.append('dessert')

    return categories if categories else ['snacks']  # Default to 'other' if no match

# Apply the mapping to the meal_time column
df['meal_time_categories'] = df['meal_time'].apply(map_meal_time)
```
