```python
# Define a mapping for regions
north_indian_regions = [
    'Punjabi', 'North Indian', 'Bihari', 'UP/Bihari', 'Kashmiri', 'Lucknowi',
    'Maharashtrian', 'Rajasthan', 'Gujarati', 'Bengali', 'Nepalese', 'North East',
    'Sindhi', 'Goan', 'Rajasthani', 'Mughlai', 'UP', 'Himachal Pradesh', 'Awadhi',
    'Odia', 'Uttarakhand', 'Agra', 'North India', 'Pan India', 'Agra, North India',
    'Madhya Pradesh', 'Fusion'
]

south_indian_regions = [
    'Karnataka', 'Andhra Pradesh', 'Tamil Nadu', 'Kerala', 'Hyderabad',
    'South Indian', 'Kerala/Tamil Nadu', 'Karnataka/Maharashtra',
    'Andhra Pradesh/Maharashtra', 'Hyderabadi', 'Andhra', 'Goa', 'South India',
    'Konkan'
]

# Map regions to "North Indian" or "South Indian"
def map_region(region):
    if region in north_indian_regions:
        return 'North Indian'
    elif region in south_indian_regions:
        return 'South Indian'
    else:
        return 'Other'

# Apply the mapping
df['region_category'] = df['region'].apply(map_region)
```

### region has been broken down to only two categories - north and south

```python
df.drop(columns=['region'], inplace=True)
df.rename(columns={'region_category': 'region'}, inplace=True)
```

```python
df['meal_time'].unique()
```

```python
# Define mappings for meal_time categories
breakfast_keywords = ['breakfast', 'Breakfast', 'Breakfast/Snack', 'breakfast/snacks', 'breakfast/lunch']
lunch_keywords = ['lunch', 'Lunch', 'Lunch/Dinner', 'Lunch/Snack']
dinner_keywords = ['dinner', 'Dinner', 'Lunch/Dinner']
snacks_keywords = ['snacks', 'Snack', 'Snacks', 'Dessert', 'Dessert/Snacks', 'Beverage/Snacks', 'Post-Workout', 'Side Dish', 'salad', 'side dish', 'beverage']

# Function to map meal_time to categories
def map_meal_time(meal_time):
    categories = []
    if isinstance(meal_time, str):  # Check if meal_time is a string
        meal_time_lower = meal_time.lower()
        if any(keyword in meal_time_lower for keyword in breakfast_keywords):
            categories.append('breakfast')
        if any(keyword in meal_time_lower for keyword in lunch_keywords):
            categories.append('lunch')
        if any(keyword in meal_time_lower for keyword in dinner_keywords):
            categories.append('dinner')
        if any(keyword in meal_time_lower for keyword in snacks_keywords):
            categories.append('snacks')
    return categories if categories else ['other']  # Default to 'other' if no match

# Apply the mapping to the meal_time column
df['meal_time_categories'] = df['meal_time'].apply(map_meal_time)
```
