```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:\calorie_RL\data\meal2_cleaned.csv')
df
```

```python
def create_diet_plan(diet_type, total_calories_required, total_budget):
    # Filter meals based on diet type
    filtered_meals = df[df['diet_type'] == diet_type].copy()

    # Sort meals by cost per serving and calories per serving (ascending order)
    filtered_meals.sort_values(by=['cost_per_serving_in_inr', 'calories_per_serving'], inplace=True)

    # Initialize variables
    diet_plan = []
    total_calories = 0
    total_cost = 0

    # Iterate through the meals to create the diet plan
    for _, meal in filtered_meals.iterrows():
        if total_calories >= total_calories_required - 100:
            break  # Stop if calorie requirement is met within the range

        # Calculate the number of servings needed to meet the calorie requirement
        remaining_calories = total_calories_required - total_calories
        servings = min(2, max(1, round(remaining_calories / meal['calories_per_serving'])))

        # Calculate the cost and calories for the selected servings
        meal_calories = servings * meal['calories_per_serving']
        meal_cost = servings * meal['cost_per_serving_in_inr']

        # Check if adding this meal exceeds the budget
        if total_cost + meal_cost > total_budget:
            continue  # Skip this meal if it exceeds the budget

        # Add the meal to the diet plan
        diet_plan.append({
            'meal_name': meal['meal_name'],
            'servings': servings,
            'calories': meal_calories,
            'cost': meal_cost
        })

        # Update total calories and cost
        total_calories += meal_calories
        total_cost += meal_cost

    # Check if the calorie requirement is met
    if total_calories < total_calories_required - 100:
        return "Unable to create a diet plan within the given budget and calorie constraints."

    # Return the diet plan as a DataFrame
    return pd.DataFrame(diet_plan)
```

```python

diet_type = 'Vegetarian'
total_calories_required = 2000
total_budget = 300

diet_plan = create_diet_plan(diet_type, total_calories_required, total_budget)
print(diet_plan)
```
